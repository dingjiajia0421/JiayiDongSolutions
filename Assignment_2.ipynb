```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df1 = pd.read_csv('Assignment 2 - USA_AL_Auburn-Opelika.AP.722284_TMY3_BASE.csv')
df2 = pd.read_csv('Assignment 2 - new.app4.csv',usecols=['time', 'W_min'])
```

```python
# Strip any leading/trailing spaces in the 'Date/Time' column
df1['Date/Time'] = df1['Date/Time'].str.strip()
df1.head()
```

```python
df2.head()
```

```python
# Convert 'Date/Time' in df1 to datetime
df1['Datetime'] = pd.to_datetime(df1['Date/Time'], format='%m/%d %H:%M:%S', errors='coerce')
df1['Datetime'] = df1['Datetime'].dt.strftime('%m/%d %H:%M')
df1.set_index('Datetime', inplace=True)
```

```python
df1.head()
```

```python
# Convert 'time' in df2 to datetime
df2['Datetime'] = pd.to_datetime(df2['time'], format='%m/%d/%Y %H:%M', errors='coerce')
df2.set_index('Datetime', inplace=True)
df2_hourly = df2.resample('H').sum().reset_index()
df2_hourly['Datetime'] = df2_hourly['Datetime'].dt.strftime('%m/%d %H:%M')
```

```python
df2_hourly.head()
```

```python
# Merge the two dataframes on the 'Datetime' column
df2_hourly.set_index('Datetime', inplace=True)
df_merged = pd.merge(df1, df2_hourly, left_index=True, right_index=True, suffixes=('_df1', '_df2'))
```

```python
# Create a new column for total hourly consumption by summing all columns
df_merged.drop(columns=['time', 'Date/Time'], inplace=True, errors='ignore')
df_merged['W_min'] = df_merged['W_min'] / 1000
df_merged['Total_Consumption_kW'] = df_merged.sum(axis=1)
```

```python
# Plotting the data to check for anomalies
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_merged['Total_Consumption_kW'])
plt.title('Hourly Electricity Consumption')
plt.xlabel('Datetime')
plt.ylabel('Electricity Consumption (kW)')
ax = plt.gca()
xticks = ax.get_xticks()
ax.set_xticks(xticks[::len(xticks)//10])
plt.xticks(rotation=45)
plt.show()
```

1. We can observe consistent high consumption periods from the start to mid-July. However, a significant drop in mid-July suggests potential events like power outages or operational changes.
2. Besides, a stabilization pattern post-mid-July with periodic spikes, and an isolated peak in September indicate an unusual event or sudden demand spike.
3. These anomalies highlight significant shifts in usage patterns, likely due to operational or external factors impacting electricity consumption. Further investigation would be necessary to determine the precise causes of these irregularities.

```python
# Check for anomalies by plotting hourly, weekday, and monthly trends
df_merged['Hour'] = df_merged.index.str[6:8].astype(int)  # Extract hour from the datetime string
df_merged['Weekday'] = pd.to_datetime(df_merged.index, format='%m/%d %H:%M').dayofweek
df_merged['Month'] = pd.to_datetime(df_merged.index, format='%m/%d %H:%M').month

plt.figure(figsize=(14, 7))
sns.boxplot(x='Hour', y='Total_Consumption_kW', data=df_merged)
plt.title('Total Hourly Electricity Consumption by Hour of the Day')
plt.show()

plt.figure(figsize=(14, 7))
sns.boxplot(x='Weekday', y='Total_Consumption_kW', data=df_merged)
plt.title('Total Hourly Electricity Consumption by Day of the Week')
plt.show()

plt.figure(figsize=(14, 7))
sns.boxplot(x='Month', y='Total_Consumption_kW', data=df_merged)
plt.title('Total Hourly Electricity Consumption by Month')
plt.show()
```

Hourly: Peak electricity consumption occurs between 11 AM and 1 PM, likely due to increased appliance use. Low consumption during early morning (1 AM - 6 AM) and late evening (8 PM - 12 AM).


Weekday: Consistent electricity consumption across all days of the week. Sporadic high consumption outliers suggest specific events on various days.

Monthly: July shows the highest variability and peaks in electricity consumption, indicating higher use of cooling devices during summer.
